```python
import matplotlib.pyplot as plt
import numpy as np

# Example data
age_groups = [
    "0-4",
    "5-9",
    "10-14",
    "15-19",
    "20-24",
    "25-29",
    "30-34",
    "35-39",
    "40-44",
    "45-49",
    "50-54",
    "55-59",
    "60-64",
    "65-69",
    "70-74",
    "75-79",
    "80+",
]
males = [
    500,
    600,
    550,
    700,
    800,
    750,
    650,
    600,
    620,
    580,
    500,
    450,
    400,
    350,
    300,
    250,
    200,
]
females = [
    480,
    590,
    530,
    680,
    780,
    740,
    620,
    580,
    600,
    560,
    490,
    440,
    390,
    340,
    290,
    240,
    190,
]

# Convert males and females to negative and positive values for plotting
males_negative = [-m for m in males]
females_positive = females

# Create the plot
fig, ax = plt.subplots(figsize=(12, 8))

# Plot the population pyramid
ax.barh(age_groups, males_negative, color="blue", label="Males", edgecolor="black")
ax.barh(age_groups, females_positive, color="red", label="Females", edgecolor="black")

# Set labels and title
ax.set_xlabel("Number of Observations")
ax.set_ylabel("Age Groups")
ax.set_title("Population Pyramid")
ax.legend()

# Add grid for better readability
ax.grid(True)

# Show plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Sample DataFrame
data = {
    "Age Group": [
        "0-4",
        "5-9",
        "10-14",
        "15-19",
        "20-24",
        "25-29",
        "30-34",
        "35-39",
        "40-44",
        "45-49",
        "50-54",
        "55-59",
        "60-64",
        "65-69",
        "70-74",
        "75-79",
        "80+",
    ],
    "Gender": [
        "F",
        "M",
        "F",
        "U",
        "M",
        "F",
        "F",
        "M",
        "U",
        "M",
        "F",
        "M",
        "F",
        "U",
        "F",
        "M",
        "F",
    ],
}

df = pd.DataFrame(data)

# Count occurrences for each age group and gender
age_gender_counts = df.groupby(["Age Group", "Gender"]).size().unstack(fill_value=0)

# Prepare data for plotting
age_groups = age_gender_counts.index
males = age_gender_counts["M"]
females = age_gender_counts["F"]

# Convert males to negative values for plotting
males_negative = -males

# Create the plot
fig, ax = plt.subplots(figsize=(12, 8))

# Plot the population pyramid
ax.barh(age_groups, males_negative, color="blue", label="Males", edgecolor="black")
ax.barh(age_groups, females, color="red", label="Females", edgecolor="black")

# Set labels and title
ax.set_xlabel("Number of Observations")
ax.set_ylabel("Age Groups")
ax.set_title("Population Pyramid")
ax.legend()

# Add grid for better readability
ax.grid(True)

# Show plot
plt.show()
```
